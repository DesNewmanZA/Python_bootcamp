# Introduction

Since Jan. 1, 2015, [The Washington Post](https://www.washingtonpost.com/) has been compiling a database of every fatal shooting in the US by a police officer in the line of duty. 

<center><img src=https://i.imgur.com/sX3K62b.png></center>

While there are many challenges regarding data collection and reporting, The Washington Post has been tracking more than a dozen details about each killing. This includes the race, age and gender of the deceased, whether the person was armed, and whether the victim was experiencing a mental-health crisis. The Washington Post has gathered this supplemental information from law enforcement websites, local new reports, social media, and by monitoring independent databases such as "Killed by police" and "Fatal Encounters". The Post has also conducted additional reporting in many cases.

There are 4 additional datasets: US census data on poverty rate, high school graduation rate, median household income, and racial demographics. [Source of census data](https://factfinder.census.gov/faces/nav/jsf/pages/community_facts.xhtml).

From the above, we see that a lot of fields that we would expect to be floats are strings. Let's fix that and then examine missings

```python
# If income contains the below, it might skew estimates - remove
df_hh_income['Median Income'] = df_hh_income['Median Income'].replace(r'[\(X\),\+\-]', np.nan, regex=True)
df_hh_income['Median Income']=df_hh_income['Median Income'].astype(float)

# If poverty rate is -, it is 0
df_pct_poverty['poverty_rate'] = df_pct_poverty['poverty_rate'].replace(r'[\-]', 0, regex=True)
df_pct_poverty['poverty_rate'] = df_pct_poverty['poverty_rate'].astype(float)

# If % completed high school is -, it is 0
df_pct_completed_hs['percent_completed_hs'] = df_pct_completed_hs['percent_completed_hs'].replace(r'[\-]', 0, regex=True)
df_pct_completed_hs['percent_completed_hs'] = df_pct_completed_hs['percent_completed_hs'].astype(float)

# Convert shares to floats
for col in df_share_race_city.filter(like='share_').columns:
    df_share_race_city[col] = df_share_race_city[col].replace(r'[\(X\)]', np.nan, regex=True)
    df_share_race_city[col] = df_share_race_city[col].astype(float)

# Ensure each entry totals 100%
share_cols = [c for c in df_share_race_city.columns if c.startswith('share_')]
df_share_race_city['share_total'] = df_share_race_city[share_cols].sum(axis=1)
df_share_race_city[share_cols] = df_share_race_city[share_cols].div(df_share_race_city['share_total'], axis=0)*100
df_share_race_city = df_share_race_city.drop('share_total', axis=1)
```

# Chart the Poverty Rate in each US State

Create a bar chart that ranks the poverty rate from highest to lowest by US state. Which state has the highest poverty rate? Which state has the lowest poverty rate?  Bar Plot

```python
poverty_by_state = df_pct_poverty[['poverty_rate', 'Geographic Area']].groupby('Geographic Area').mean()
poverty_by_state = poverty_by_state.sort_values(by='poverty_rate', ascending=False)

plt.figure(figsize=(16, 7))
sns.barplot(data=poverty_by_state, x='Geographic Area', y='poverty_rate')
plt.title('Poverty rate by state')
plt.ylabel('Poverty rate')
plt.show()
```

##### Mississippi has the highest poverty rate while New Jersey has the lowest

# Chart the High School Graduation Rate by US State

Show the High School Graduation Rate in ascending order of US States. Which state has the lowest high school graduation rate? Which state has the highest?

```python
hs_by_state = df_pct_completed_hs[['percent_completed_hs', 'Geographic Area']].groupby('Geographic Area').mean()
hs_by_state = hs_by_state.sort_values(by='percent_completed_hs', ascending=False)

plt.figure(figsize=(16, 7))
sns.barplot(data=hs_by_state, x='Geographic Area', y='percent_completed_hs')
plt.title('High school completion rate by state')
plt.ylabel('High school completion rate')
plt.show()
```

##### Massachusetts has the highest rate whereas Texas has the lowest 

# Visualise the Relationship between Poverty Rates and High School Graduation Rates

#### Create a line chart with two y-axes to show if the ratios of poverty and high school graduation move together.  

```python
fig, ax1 = plt.subplots(figsize=(16, 7))

sns.lineplot(x='Geographic Area', y='poverty_rate', data=poverty_by_state, ax=ax1, label='Poverty rate', color='red')
ax1.set_ylabel('Poverty rate')

ax2 = ax1.twinx()
sns.lineplot(x='Geographic Area', y='percent_completed_hs', ax=ax2, data=hs_by_state, label='High school completion rate', color='green')
ax2.set_ylabel('High school completion rate')

plt.title('Poverty and high school completetion rate by state')
plt.xlabel('State')
plt.legend()

plt.show()
```

##### There is an inverse relationship between poverty rate and high school completion rate

#### Now use a Seaborn .jointplot() with a Kernel Density Estimate (KDE) and/or scatter plot to visualise the same relationship

```python
joint_df = poverty_by_state.merge(hs_by_state, on='Geographic Area')
joint_df.sort_values('percent_completed_hs', ascending=False, inplace=True)
```

```python
sns.jointplot(data=joint_df, x='poverty_rate', y='percent_completed_hs', kind='kde', fill=True, height=5, ratio=5)
plt.xlabel("High School Graduation Rate")
plt.ylabel("Poverty Rate")
plt.show()
```

#### Seaborn's `.lmplot()` or `.regplot()` to show a linear regression between the poverty ratio and the high school graduation ratio. 

```python
sns.regplot(joint_df, x="percent_completed_hs", y="poverty_rate")

plt.xlabel("High School Graduation Rate")
plt.ylabel("Poverty Rate")
plt.title('Linear regression of poverty and high school completion rates')

plt.show()
```

# Rate of Death by Race

Find the share of each race in the top 10 cities. Contrast this with the top 10 cities of police killings to work out the rate at which people are killed by race for each city. 

```python
merged_df = top10_cities.merge(df_fatalities, on=['state', 'city'])
merged_df = merged_df.groupby(['state', 'city', 'count'])['race'].value_counts(dropna=False).reset_index(name='death_rate')
merged_df['death_rate'] = (merged_df['death_rate'] / merged_df['count']) * 100
merged_df = merged_df[merged_df['race'].isin(["A", "B", "H", "N", "W"])]
merged_df
```

```python
df_share_race_city.rename(columns={'City': 'city', 'Geographic area': 'state'}, inplace=True)
df_share_race_city['city'] = df_share_race_city['city'].str.replace(' City', '', regex=False)
df_share_race_city['city'] = df_share_race_city['city'].str.replace(' city', '', regex=False)
df_share_race_city = df_share_race_city.melt(id_vars=["state", "city"], var_name="race", value_name="race_share")
df_share_race_city
```

```python
df_share_race_city['race'] = df_share_race_city['race'].str.replace('share_white', 'W')
df_share_race_city['race'] = df_share_race_city['race'].str.replace('share_black', 'B')
df_share_race_city['race'] = df_share_race_city['race'].str.replace('share_native_american', 'N')
df_share_race_city['race'] = df_share_race_city['race'].str.replace('share_hispanic', 'H')
df_share_race_city['race'] = df_share_race_city['race'].str.replace('share_asian', 'A')
df_share_race_city
```

```python
top10_cities_race = pd.merge(merged_df, df_share_race_city, on=['city', 'state', 'race'], how='right')
unique_cities = top10_cities_race[top10_cities_race.death_rate.notna()]['city'].unique().tolist()
top10_cities_race = top10_cities_race[top10_cities_race.city.isin(unique_cities)]
top10_cities_race['death_rate'] = top10_cities_race['death_rate'].fillna(0)
top10_cities_race = top10_cities_race[top10_cities_race['count'].notna()]
top10_cities_race = top10_cities_race.sort_values(by=['city', 'death_rate'], ascending=[True, False])
top10_cities_race
```

```python
for city in top10_cities['city']:
    fig, ax = plt.subplots()
    city_info = top10_cities_race[top10_cities_race['city'] == city]

    sns.barplot(data=city_info, x='race', y='race_share', ax=ax)
    ax.set_xlabel("Race")
    ax.set_ylabel("Proportion")
    
    ax2 = ax.twinx()
    sns.lineplot(data=city_info, x='race', y='death_rate', ax=ax2, color='black', errorbar=None)
    ax2.set_ylabel("Death rate by race")
    ax2.grid(False)
    
    plt.title(f'City: {city}')
    plt.show()
```

# Number of Police Killings Over Time

Analyse the Number of Police Killings over Time. Is there a trend in the data? 

```python
df_fatalities['date'] = pd.to_datetime(df_fatalities['date'])
deaths_over_time = df_fatalities.groupby('date').size().reset_index(name='count').sort_values('date')
deaths_over_time['year'] = deaths_over_time['date'].dt.year
deaths_over_time['month'] = deaths_over_time['date'].dt.month
deaths_over_time = deaths_over_time.groupby(['year', 'month'])['count'].sum().reset_index()
deaths_over_time['year_month'] = pd.to_datetime(deaths_over_time[['year', 'month']].assign(day=1))
deaths_over_time
```

```python
plt.figure(figsize=(20, 6))

plt.plot(deaths_over_time.year_month, deaths_over_time['count'])
plt.xlabel('Date')
plt.ylabel('Number of deaths')
plt.title('Number of police killings over time')

plt.show()
```
