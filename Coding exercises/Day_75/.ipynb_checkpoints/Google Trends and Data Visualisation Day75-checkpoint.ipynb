### Converting from Daily to Monthly Data

# Data Visualisation

### Tesla Stock Price v.s. Search Volume

**Challenge:** Plot the Tesla stock price against the Tesla search volume using a line chart and two different axes. Label one axis 'TSLA Stock Price' and the other 'Search Trend'. 

```python
import matplotlib.dates as mdates

years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')

plt.figure(figsize=(14, 8), dpi=150)
plt.title("Tesla Web Search vs Price", fontsize=20)
plt.xticks(fontsize=14, rotation=45)

ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

ax1 = plt.gca() 
ax2 = ax1.twinx()
ax1.set_ylabel('TSLA Stock Price', color='#bf281d')
ax2.set_ylabel('Search Trend', color='cornflowerblue')
ax1.set_xlim(ax1.set_xlim([df_tesla["MONTH"].min(), df_tesla["MONTH"].max()]))
    
ax1.plot(df_tesla["MONTH"], df_tesla["TSLA_USD_CLOSE"], label = "Share price", color='#bf281d', linewidth=5)
ax2.plot(df_tesla["MONTH"], df_tesla["TSLA_WEB_SEARCH"], label = "Number of searches", color='cornflowerblue', linewidth=5)

plt.show()
```

```python
import matplotlib.dates as mdates

years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')

plt.figure(figsize=(14, 8), dpi=150)
plt.title("Bitcoin News Search vs Resampled Price", fontsize=20)
plt.xticks(fontsize=14, rotation=45)

ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

ax1 = plt.gca() 
ax2 = ax1.twinx()
ax1.set_ylabel('BTC Price', color='#bf281d')
ax2.set_ylabel('Search Trend', color='cornflowerblue')
ax1.set_xlim(ax1.set_xlim([df_btc_search["MONTH"].min(), df_btc_search["MONTH"].max()]))
    
ax1.plot(df_btc_monthly["DATE"], df_btc_monthly["CLOSE"], label = "Bitcoin price", color='#bf281d', linewidth=2, linestyle='dashed')
ax2.plot(df_btc_search["MONTH"], df_btc_search["BTC_NEWS_SEARCH"], label = "Number of searches", color='cornflowerblue', linewidth=2, marker='o')

plt.show()
```

### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

**Challenge** Plot the search for "unemployment benefits" against the unemployment rate. 
1. Change the title to: Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate <br>
2. Change the y-axis label to: FRED U/E Rate <br>
3. Change the axis limits <br>
4. Add a grey [grid](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.grid.html) to the chart to better see the years and the U/E rate values. Use dashes for the line style<br> 
5. Can you discern any seasonality in the searches? Is there a pattern? 

```python
plt.figure(figsize=(14, 8), dpi=150)
plt.title("Unemployment rate vs unemployment benefits web searches", fontsize=20)
plt.xticks(fontsize=14, rotation=45)
plt.grid(color='grey', linestyle='--', linewidth=1)

ax1 = plt.gca() 
ax2 = ax1.twinx()
ax1.set_ylabel('Unemployment rate', color='#bf281d')
ax2.set_ylabel('Unemployment benefits search trned', color='cornflowerblue')
ax1.set_xlim(ax1.set_xlim([df_unemployment["MONTH"].min(), df_unemployment["MONTH"].max()]))
    
ax1.plot(df_unemployment["MONTH"], df_unemployment["UNRATE"], label = "Unemployment rate", color='#bf281d', linewidth=2, linestyle='dashed')
ax2.plot(df_unemployment["MONTH"], df_unemployment["UE_BENEFITS_WEB_SEARCH"], label = "Unemployment benefit searches", color='cornflowerblue', linewidth=2, marker='o')

plt.show()
```

**Challenge**: Calculate the 3-month or 6-month rolling average for the web searches. Plot the 6-month rolling average search data against the actual unemployment. What do you see in the chart? Which line moves first?


```python
roll_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()

plt.figure(figsize=(14, 8), dpi=150)
plt.title("Unemployment rate vs unemployment benefits web searches", fontsize=20)
plt.xticks(fontsize=14, rotation=45)
plt.grid(color='grey', linestyle='--', linewidth=1)

ax1 = plt.gca() 
ax2 = ax1.twinx()
ax1.set_ylabel('Unemployment rate', color='#bf281d')
ax2.set_ylabel('Unemployment benefits search trned', color='cornflowerblue')
ax1.set_xlim(ax1.set_xlim([df_unemployment["MONTH"].min(), df_unemployment["MONTH"].max()]))
    
ax1.plot(df_unemployment["MONTH"], roll_df["UNRATE"], label = "Unemployment rate", color='#bf281d', linewidth=2, linestyle='dashed')
ax2.plot(df_unemployment["MONTH"], roll_df["UE_BENEFITS_WEB_SEARCH"], label = "Unemployment benefit searches", color='cornflowerblue', linewidth=2, marker='o')

plt.show()
```
